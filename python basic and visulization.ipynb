```python
#Create a small Dataframe
import pandas as pd 
df = pd.DataFrame(index=['Atiya', 'Abbas', 'Cornelia', 'Stephanie', 'Monte'],
                  data={'Apples':[20, 10, 40, 20, 50],'Oranges':[35, 40, 25, 19, 33]})
```

```python
df
```

```python
#color=['.2',".7"]
ax = df.plot.bar(color=['.7','.3'], figsize=(16,4))
#ax.get_figure()
```

```python
#Density plot
ax = df.plot.kde(color=['.2',".7"], figsize=(16,4))
```

```python
#Let's plot a line plot, scatter plot, and bar plot in a single figure
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
fig.suptitle('Two Variable Plots')
df.plot.line(ax=ax1, title='Line plot')
df.plot.scatter(x='Apples', y='Oranges',ax=ax2, title='Scatterplot')
df.plot.bar(ax=ax3, title='Bar plot')
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
```

```python
tips
```

```python
#strip plot
sns.stripplot(y= 'tip', data=tips,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Grouping with stripplot
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex', orient='h')
plt.xlabel('tip ($)')
plt.show()
```

```python
#plt.subplots(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Joint plot
sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
#Density plot
sns.jointplot(x= 'total_bill', y= 'tip', data=tips,kind='kde')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue= 'sex')
plt.show()
```
